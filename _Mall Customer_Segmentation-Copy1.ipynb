```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Mall_Customers.csv")

df.shape
```

```python
df.head()
```

```python
x=df.loc[:,['Annual Income (k$)',
             'Spending Score (1-100)']].values

print(x.shape)
```

```python
from sklearn.cluster import KMeans


wcss=[]

for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
    

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method',fontsize=20)
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
    
```

```python
from sklearn.cluster import KMeans

km=KMeans(n_clusters=5,init='k-means++',max_iter=300,random_state=0)
y_means=km.fit_predict(x)


plt.rcParams['figure.figsize']=(15,10)
plt.style.use('fivethirtyeight')

plt.scatter(x[y_means==0,0],x[y_means==0,1],s=100 , c='pink',label='general')
plt.scatter(x[y_means==1,0],x[y_means==1,1],s=100 , c='yellow',label='spendthrift')
plt.scatter(x[y_means==2,0],x[y_means==2,1],s=100 , c='black',label='target')
plt.scatter(x[y_means==3,0],x[y_means==3,1],s=100 , c='magenta',label='careful')
plt.scatter(x[y_means==4,0],x[y_means==4,1],s=100 , c='orange',label='miser')


plt.style.use('fivethirtyeight')
plt.title('K means Clustring',fontsize=20)
plt.xlabel('Annual income')
plt.ylabel('spending Score')
plt.legend()
plt.grid()
plt.show()

```

```python
from sklearn.metrics import silhouette_score

score=silhouette_score(x,km.labels_,metric='euclidean')

print('silhouette score', score)
```

```python
x=df.loc[:,['Annual Income (k$)',
             'Spending Score (1-100)']].values

print(x.shape)
```

```python
import scipy.cluster.hierarchy as sch

plt.rcParams['figure.figsize']==(4,2)

dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram ' , fontsize=20)
plt.xlabel('Customers')
plt.ylabel('Ecuclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')

y_hc=hc.fit_predict(x)
```

```python


plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100 , c='pink',label='general')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100 , c='yellow',label='spendthrift')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100 , c='black',label='target')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100 , c='magenta',label='careful')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100 , c='orange',label='miser')


plt.style.use('default')
plt.rcParams['figure.figsize']=(15,5)
plt.title('Hierarchial Clustring',fontsize=20)
plt.xlabel('Annual income')
plt.ylabel('spending Score')
plt.legend()
plt.grid()
plt.show()
```

```python
x=df.loc[:,['Annual Income (k$)',
             'Spending Score (1-100)']].values

print(x.shape)
```

```python
from sklearn.neighbors import NearestNeighbors

nigh=NearestNeighbors(n_neighbors=2)
nbrs=nigh.fit(x)
distances,indices =nbrs.kneighbors(x)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.rcParams['figure.figsize']=(5,3)
plt.plot(distances)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan=DBSCAN(eps=8,min_samples=4)
clusters=dbscan.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=clusters,cmap='plasma')
plt.xlabel("income")
plt.ylabel("Spending Score ")
plt.show()
```
